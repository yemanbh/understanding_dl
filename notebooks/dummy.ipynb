```python
import pyvista as pv
import numpy as np

# Create a grid of points
x, y, z = np.meshgrid(np.linspace(-5, 5, 20),
                      np.linspace(-5, 5, 20),
                      np.linspace(-5, 5, 20))

# Define vectors (e.g., a simple magnetic field-like pattern)
u = -y
v = x
w = z * 0.1

# Create a PyVista PolyData object
grid = pv.StructuredGrid(x, y, z)
grid["vectors"] = np.c_[u.ravel(), v.ravel(), w.ravel()]

# Create glyphs (arrows)
arrows = grid.glyph(orient="vectors", scale=False, factor=0.5)

# Plotting
plotter = pv.Plotter()
plotter.add_mesh(arrows, color="cyan")
plotter.show()
```
